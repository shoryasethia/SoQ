```python
from matplotlib import pyplot as plt
from statsmodels.graphics import tsaplots
import pandas as pd
```

```python
df = pd.read_excel('data/earthquake_data.xlsx')
df
```

```python
df.set_index('Year')
```

```python
plt.figure(figsize=(10,8))
plt.scatter(df.index,df['Quakes'])
plt.plot(df.index,df['Quakes'])
plt.show
```

```python
tsaplots.plot_pacf(df['Quakes'], lags=10)
```

## So lag 1 and 3 indicates that AR(1) or AR(3) might we useful to predict the next eqarthquake

### The blue shaded region is a region between plus-minus 1.96/sqrt(105). 105 is n : number of observations
### This is 5% significant level, alpha = 0.005, Z_(1-alpha/2) = 1.96
### The spike lying above it are of our interest
