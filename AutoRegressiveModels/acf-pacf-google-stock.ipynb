```python
from matplotlib import pyplot as plt
from statsmodels.graphics import tsaplots
import pandas as pd
```

```python
df = pd.read_excel('data/google_stock.xlsx')
df
```

```python
df.set_index('date', inplace=True)
df
```

```python
plt.figure(figsize=(15,7))
plt.scatter(df.index,df['price'])
plt.xticks(rotation = 90)
plt.xlabel('date')
plt.ylabel('closing price')
plt.show()
```

```python
tsaplots.plot_acf(df, lags=10)
```

## Ignore lag = 0; The autocorrelation function starts a lag 0, which is the correlation of the time series with itself and therefore results in a correlation of 1.

```python
tsaplots.plot_pacf(df, lags = 10)
```
