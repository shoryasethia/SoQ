```python
from statsmodels.tsa.stattools import adfuller
import pandas as pd
```

```python
df1 = pd.read_excel('data/google_stock.xlsx')
df2 = pd.read_excel('data/Stationary_Time_Series_Temperature_Data.xlsx')
df3 = pd.read_excel('data/earthquake_data.xlsx')
```

```python
import matplotlib.pyplot as plt

def adfuller_test(series):
  result = adfuller(series)
  print(f"p-value = {result[1]}")

  if result[1] < 0.05:
    print('Time Series is Stationary.')
    plt.plot(series)
    plt.show()

  else :
    print('Time Series is Non Statioanry.')
    plt.plot(series)
    plt.show()
```

```python
adfuller_test(df1['price'])
```

```python
adfuller_test(df2['Temperature (C)'])
```

```python
adfuller_test(df3['Quakes'])
```
